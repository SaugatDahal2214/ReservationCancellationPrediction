```python
df.hist(bins=20, figsize=(20,15))
plt.suptitle("Distribution of numerical feature", fontsize=16)
plt.savefig('numerical_distribution.jpg')
plt.show()
```

```python
booking_status_counts = df['booking_status'].value_counts()
```

```python
#Booking status
plt.figure(figsize=(8, 8))
plt.pie(booking_status_counts, labels=booking_status_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
plt.title('Booking Status Distribution')
plt.savefig("Booking_distribution_piechart.png")
plt.show()
```

```python
#Categorical features distributions by booking status

categorical_colms = ["type_of_meal_plan", "room_type_reserved"]

for col in categorical_colms:
    plt.figure(figsize=(10,10))
    sns.countplot(x=col, hue='booking_status', data=df)
    plt.title(f'{col} Distributed by Booking Status')
    plt.xticks(rotation=45)
    plt.show()
```

```python
numerical_cols = ['lead_time', 'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled', 'avg_price_per_room']

for cols in numerical_cols:
    plt.figure(figsize=(10, 10))
    sns.boxplot(x='booking_status', y=cols, data=df)
    plt.title(f'{cols} vs Booking Status')
    plt.show()
```

```python
#Distribution of lead time and average price

sns.histplot(df['lead_time'], kde=True)
plt.title('Distribution of lead time')
plt.savefig("Distribution of lead time")
plt.show()

sns.histplot(df['avg_price_per_room'], kde=True)
plt.title('Distribution of Previous Cancellation')
plt.savefig("Distribution of Previous Cancellation")
plt.show()
```

```python
cap_threshold = 200
df['lead_time'] = df['lead_time'].apply(lambda x: min(x, cap_threshold))
```

```python
# Calculate Q1 and Q3 for lead_time
Q1 = df['avg_price_per_room'].quantile(0.25)
Q3 = df['avg_price_per_room'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df = df[(df['avg_price_per_room'] >= lower_bound) & (df['avg_price_per_room'] <= upper_bound)]

# Check the result
print(df['avg_price_per_room'].describe())
```

```python
df.shape
```

```python
cap_threshold = 180
df['avg_price_per_room'] = df['avg_price_per_room'].apply(lambda x: min(x, cap_threshold))
```

```python
numerical_cols = ['lead_time', 'avg_price_per_room']

for cols in numerical_cols:
    plt.figure(figsize=(10, 10))
    sns.boxplot(x='booking_status', y=cols, data=df)
    plt.title(f'{cols} vs Booking Status')
    plt.show()
```

ENCODING THE CATEGORICAL VALUE

```python
df['type_of_meal_plan'].unique()
```

```python
meal_plan_encoding = {
    'Not Selected':0,
    'Meal Plan 1' : 1,
    'Meal Plan 2' : 2,
}


df['type_of_meal_plan'] = df['type_of_meal_plan'].map(meal_plan_encoding)
```

```python
df['type_of_meal_plan'].unique()
```

```python
df['room_type_reserved'].unique()
```

```python
room_type_encoding = {
    'Room_Type 1': 0,
    'Room_Type 2': 1,
    'Room_Type 3': 2,
    'Room_Type 4': 3,
    'Room_Type 5': 4,
    'Room_Type 6': 5,
    'Room_Type 7': 6,
}

df['room_type_reserved'] = df['room_type_reserved'].map(room_type_encoding)
```

```python
df['room_type_reserved'].unique()
```

```python
df['booking_status'].unique()
```

```python
Booking_encode ={
    'Not_Canceled':0,
    'Canceled':1
}

df['booking_status'] = df['booking_status'].map(Booking_encode)
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.savefig("Correlation_Matrix.jpg")
plt.show()
```
